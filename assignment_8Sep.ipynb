```python
import seaborn as sns
import numpy as np
import matplotlib as plt
```

```python
df=sns.load_dataset('car_crashes')
```

```python
ndf = df.select_dtypes(include=[np.number])
ndf.head()
```

```python
cm = ndf.corr()
tcf = cm.index
g=sns.heatmap(ndf[tcf].corr(),annot=True,cmap="RdYlGn")
```

There is a strong positive correlation between (total, speeding, alcohol, not_distracted, no_previous) and there exists strong correlation between ins_premium and ins_losses. The rest are very weak negative correlation.

```python
sns.barplot(x='abbrev', y='ins_premium', data=df)
```

The states with high insurance have a higher risk of crashes occuring. The state with lower insurance has a lower risk of car crashing. The average insurance across USA is 866 dollars but four states have insurance higher than 1200. These states are Washington D.C, Lousiana, New Jersey and New York. In these states there is a chance that some drivers do not get insurance for their cars. Idaho, Iowa, Maine, North Dakota, Ohio, South Dakota and Wisconsin have insurance lower than 700. More drivers would be taking insurance in these states than the above four.

```python
sns.barplot(x='abbrev', y='ins_losses', data=df)
```

The average loss incurred by insurance companies across USA is 134 but three states have loss higher than 175. These states are Lousiana, Maryland and Oklahoma. Lousiana has both high insurance and high losses. Idaho, Maine, Montana and South Dakota have losses lesser than 100. Other than Montana all of them have low insurance. This supports the assumption that these states have low risk of car accidents.

```python
sns.scatterplot(x='ins_premium', y='ins_losses', data=df)
```

This shows a linear relationship between insurance cost and loss incurred by company. Higher the loss incurred, higher the insurance price in the state.

```python
sns.barplot(x='abbrev', y='total', data=df)
```

Mean number of car accidents per billion kilometers across USA is 15. Arkansas, Kentucky, Lousiana, Montana, North Dakota, South Carolina and West Virginia have mean number of accidents higher than 20 and Washington D.C, Massachusets and Minnesota have mean number lower than 10. Of these Montana has high number of car crashes but low loss incurred. Washington D.C has low number of accidents and high cost of insurance. And most of these states have average loss incurred and average cost of insurance but show high numbers in this metric. This means that total number of crashes is not a good metric for measuring risk factor.

```python
d1=df[['abbrev', 'total', 'speeding', 'alcohol', 'not_distracted', 'no_previous']].sort_values(by='total').tail(10)
d1
```

```python
sns.pointplot(x='abbrev', y='speeding', data=d1, color='red')
sns.pointplot(x='abbrev', y='alcohol', data=d1, color='green')
sns.pointplot(x='abbrev', y='not_distracted', data=d1, color='blue')
sns.pointplot(x='abbrev', y='no_previous', data=d1, color='black')
```

In the above graph, we can see that among the states with least number of accidents, we can see that the line for previous accidents and alcohol differ. This means that drunk driving could lead to repeated accidents.
